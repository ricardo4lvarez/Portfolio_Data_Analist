# Feature Engineering

## Variables num√©ricas

Debido a que la variable dependiente es de 0 o 1, o Verdadero o Falso la mejor prueba estad√≠stica para corroborar su relaci√≥n significativa es **T Student de dos colas** para variables independientes num√©ricas.

```python
from scipy.stats import ttest_ind
```

Utilizando la divisi√≥n previamente hecha del df entre gente que abandona y gente que no, vamos a hacer el test.

```python
results = []

for var in num_vars:
    t_stat, p_value = ttest_ind(grupo1[var], grupo2[var], nan_policy='omit')
    results.append({"variable": var, "t_stat": t_stat, "p_value": p_value})

results_df = pd.DataFrame(results).sort_values("p_value")
results_df["Significancia"] = np.where(results_df["p_value"] < 0.05, "Significativa", "No significativa")
print(results_df)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=results_df, x="p_value", y="variable", palette="viridis", edgecolor="black", hue="p_value")
plt.axvline(x=0.05, color='red', linestyle='--', label='p = 0.05')

plt.title("P-valores del Test t por Variable")
plt.xlabel("P-valor")
plt.ylabel("Variable")
plt.legend()

plt.tight_layout()
plt.show()
```

Tanto gr√°ficamente como en el "data frame" que imprim√≠ podemos ver las variables que son o no significativas, en este caso el t_stat nos indica que hay diferencia entre sus promedios y el p_value si es o no debido al azar, en este caso lo establec√≠ en 0.05 o 5%.

En un momento lo vamos a ver gr√°ficamente.

```python
mean_values = {
    "Balance": new_df.groupby("Churn")["Balance"].mean().reset_index(),
    "CreditScore": new_df.groupby("Churn")["CreditScore"].mean().reset_index(),
    "Age": new_df.groupby("Churn")["Age"].mean().reset_index(),
}

vars_to_plot = ["Balance", "CreditScore", "Age"]
colors = ['#bee7e8', '#bf4342']

fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 12))
fig.tight_layout(pad=4)

for i, var in enumerate(vars_to_plot):
    sns.boxplot(data=new_df, x="Churn", y=var, hue="Churn", palette=colors, ax=axes[i, 0], legend=False)
    axes[i, 0].set_title(f"{var} por Churn")

    ax = sns.barplot(data=mean_values[var], x="Churn", y=var, hue="Churn", palette=colors, ax=axes[i, 1], legend=False)
    axes[i, 1].set_title(f"{var} promedio por Churn")
    for container in ax.containers:
        ax.bar_label(container)

plt.show()
```

## Variables Categ√≥ricas

En este caso voy a utilizar el test chi-cuadrado para analizar la relaci√≥n.

```python
from scipy.stats import chi2_contingency
```

```python
cat_vars = new_df.select_dtypes("category")
results = []

for var in cat_vars:
    tabla_contingencia = pd.crosstab(new_df[var], new_df["Churn"])
    chi2, p, dof, expected = chi2_contingency(tabla_contingencia)
    results.append({"Variable": var, "Chi2": chi2, "p_value": p})

chi_results = pd.DataFrame(results).sort_values("p_value")
chi_results["Significancia"] = np.where(chi_results["p_value"] < 0.05, "Significativa", "No significativa")
print(chi_results)
```

# Machine Learning

## Entrenamiento

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import SMOTE
from sklearn.metrics import make_scorer, f1_score, accuracy_score, precision_score, recall_score
```

```python
X = df_features.drop("Churn", axis=1)
y = df_features["Churn"]

X = pd.get_dummies(X, drop_first=True)

# Aqui elimino algunas de las columnas que resultaron en menos de 0.01 de influencia en el modelo.
X.drop(columns=["Edades_90+", "Edades_81-90"], axis=1, inplace=True)
```

```python
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=10),
    "Random Forest": RandomForestClassifier(random_state=10),
    "KNN": KNeighborsClassifier(),
    "XGBoost": XGBClassifier(eval_metric='logloss', random_state=10)
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=10)
scoring = {
    'Accuracy': make_scorer(accuracy_score),
    'Precision': make_scorer(precision_score),
    'Recall': make_scorer(recall_score),
    'F1': make_scorer(f1_score)
}
```

```python
results = []

for name, model in models.items():
    pipeline = ImbPipeline(steps=[('scaler', StandardScaler()), ('smote', SMOTE(random_state=10)), ('classifier', model)])

    row = {"Model": name}
    for metric_name, scorer in scoring.items():
        scores = cross_val_score(pipeline, X, y, cv=cv, scoring=scorer)
        row[metric_name] = np.mean(scores)
    results.append(row)
```

```python

results_df = pd.DataFrame(results)
print("\nResultados de la validaci√≥n cruzada:")
print(results_df)

plt.figure(figsize=(10, 6))
results_melted = results_df.melt(id_vars="Model", var_name="Metric", value_name="Score")
sns.barplot(data=results_melted, x="Model", y="Score", hue="Metric")
plt.title("Comparaci√≥n de Modelos - M√©tricas Promedio (5-Fold)")
plt.xticks(rotation=45)
plt.ylim(0, 1)
plt.tight_layout()
plt.show()
```

## Selecci√≥n del modelo

```python
modelo_final = ImbPipeline(steps=[('scaler', StandardScaler()), ('smote', SMOTE(random_state=42)), ('classifier', LogisticRegression(random_state=42))])
modelo_final.fit(X, y)
```
